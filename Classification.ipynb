```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
import numpy as np
import warnings
```

```python
rb_df = pd.read_csv('RBData_Class.csv', index_col=0)
warnings.filterwarnings("ignore")
```

```python
rb_df.head()
```

```python
current_rb_df = pd.read_csv('Current_RBData.csv', index_col=0)
current_rb_df["Seasons"] = 12 - current_rb_df.isna().sum(axis=1)

current_rb_df = current_rb_df.fillna(0) \
                    .drop(['Position', 'AV', 'First Year', 'Last Year'], axis=1)
```

```python
current_rb_df.head()
```

```python
preds = []
for rb in current_rb_df.iterrows():
    n = rb[1]["Seasons"]
    df = rb_df.drop(columns=rb_df.columns[n+1:17])
    X = df.drop(["Name", "Class"], axis=1)
    y = df["Class"]
    
    model = LogisticRegression(max_iter=1000).fit(X, y)
    curr_rb = np.array(rb[1].iloc[1:n+1]).reshape(1, -1)
    prediction = model.predict(curr_rb)
    preds.append(prediction[0])
    
current_rb_df["Prediction"] = preds
```
